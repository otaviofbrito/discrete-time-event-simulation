```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import os
```

```python
pasta_csv = '../c_project/data/'
arquivos_csv = [f for f in os.listdir(pasta_csv) if f.endswith('.csv')]

dataframes = []
for arquivo in arquivos_csv:
    caminho_arquivo = os.path.join(pasta_csv, arquivo)
    df = pd.read_csv(caminho_arquivo)
    if not df.empty:
        # Extract the number from the file name
        number = arquivo.split('output')[-1].split('.')[0]
        params = f"(E:{df.iloc[0]['Chegada']}, S:{df.iloc[0]['Saida']})"  
        df['Fonte'] = f'{number}% - {params}' 
    dataframes.append(df)

df_combined = pd.concat(dataframes, ignore_index=True)
```
